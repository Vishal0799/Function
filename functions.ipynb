**Q1. What is the difference between a function and a method in Python?**

  In Python, both functions and methods are blocks of code that can be executed multiple times from different parts of your program. However, there are key differences between them:

**Functions**
- A function is a self-contained block of code that performs a specific task.
- It is not associated with any class or object.
- Functions are defined using the def keyword.

**Methods**
- A method is a block of code that is associated with a class or object.
- It is used to perform actions on an object or to modify its state.
- Methods are also defined using the def keyword, but they are inside a class definition.

**Example:**

# Function
def greet(name):

    print(f"Hello, {name}!")

# Method
class Person:

    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}!")

**Q2. Explain the concept of function arguments and parameters in Python.**

In Python, when defining a function, you can specify arguments that the function can accept when it's called. These arguments are also known as parameters.

**Function Parameters:**

- Parameters are the variables defined in the function definition.
- They are used to receive the input values passed to the function.
- Parameters are local to the function and are not accessible outside the function.

**Function Arguments:**

- Arguments are the actual values passed to the function when it's called.
- They are assigned to the corresponding parameters in the function definition.
- Arguments can be literals, variables, or expressions.

**Example:**

def greet(name, age):

    print(f"Hello, {name}! You are {age} years old.")

 **Calling the function with arguments**

greet("John", 30)

**Types of Function Arguments:**

- **Positional arguments:** These are arguments that are passed to the function in the same order as the parameters are defined. (e.g., greet("John", 30))
- **Keyword arguments:** These are arguments that are passed to the function using the parameter name followed by an equals sign and the value. (e.g., greet(name="John", age=30))
- **Default arguments:** These are arguments that have a default value assigned to them in the function definition. (e.g., def greet(name, age=30): ...)
- **Variable-length arguments:** These are arguments that can be passed to the function in a variable number of quantities. (e.g., def greet(name, *args): ...)

**Q5. What are iterators in Python and how do they differ from iterables?**

In Python, iterators and iterables are two related but distinct concepts.

**Iterables:**

An iterable is an object that can be iterated over, meaning it can be looped through or have its elements accessed one at a time. Examples of iterables include:

- Lists: [1, 2, 3]
- Tuples: (1, 2, 3)
- Dictionaries: {"a": 1, "b": 2}
- Sets: {1, 2, 3}
- Strings: "hello"

**Iterators:**

An iterator is an object that keeps track of its position in an iterable and yields the next element each time it is called.

 Iterators are created from iterables using the iter() function.

**Here's an example:**

my_list = [1, 2, 3]
my_iter = iter(my_list)

print(next(my_iter))  # Output: 1

print(next(my_iter))  # Output: 2

print(next(my_iter))  # Output: 3

**Key differences between iterators and iterables:**

- Iterables are the objects that can be iterated over, while iterators are the objects that do the iteration.
- Iterables can be iterated over multiple times, while iterators can only be iterated over once.
- Iterators keep track of their position in the iterable, while iterables do not.

**Q6. Explain the concept of generators in Python and how they are defined.**

**Generators:**

**Definition:** Generators are a simple way to create iterators using functions and the yield keyword instead of returning values.

**Key Points:**

yield Keyword:

**Yield vs. Return:** Unlike the return statement, which exits the function, yield returns a value and pauses the functionâ€™s state, allowing it to resume from where it left off.

    for i in range(10):

    print(value)

**Q7. What are the advantages of using generators over regular functions?**

Generators have several advantages over regular functions:

1. **Memory Efficiency:** Generators use significantly less memory than regular functions because they don't store all the values in memory at once. Instead, they generate values on-the-fly, which makes them ideal for handling large datasets.

2. **Lazy Evaluation:** Generators only compute values when they are needed, which is known as lazy evaluation. This approach can lead to significant performance improvements, especially when working with expensive computations.

3. **Flexibility:** Generators can be used to create complex, dynamic sequences that would be difficult or impossible to create with regular functions.

4. **Improved Performance:** Generators can improve performance by avoiding the need to create and store large data structures. This is especially important when working with large datasets or expensive computations.

5. **Easier Handling of Infinite Sequences:** Generators make it easy to handle infinite sequences, which would be impossible to handle with regular functions.

6. **Simplified Code:** Generators can simplify code by eliminating the need for explicit loops and conditional statements.

7. **Better Support for Cooperative Multitasking:** Generators provide better support for cooperative multitasking, which allows multiple tasks to run concurrently without the need for threads or processes.

**Here's an example that demonstrates the memory efficiency of generators:**

import sys

**Create a list of numbers from 1 to 1 million**


numbers_list = list(range(1, 1000001))
print(f"Memory usage of list: {sys.getsizeof(numbers_list)} bytes")

 **Create a generator that generates numbers from 1 to 1 million**

print(f"Memory usage of generator: {sys.getsizeof(numbers_generator)} bytes")

**Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

map(), reduce(), and filter() are three fundamental functions in Python that are used for data processing and transformation. Here's a brief overview of each function, along with examples:

**1. map() function:**

The map() function applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a list of the results.

**Syntax:** map(function, iterable)

**Example:**

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

**2. reduce() function:**

The reduce() function applies a given function to the items of an iterable, going from left to right, so as to reduce the iterable to a single output.

**Syntax:** reduce(function, iterable)

**Example:**

import functools
numbers = [1, 2, 3, 4, 5]
sum_of_numbers = functools.reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)  # Output: 15

Note: In Python 3.x, reduce() is moved to the functools module.

**3. filter() function:**

The filter() function constructs an iterator from elements of an iterable for which a function returns true.

**Syntax:** filter(function, iterable)

**Example:**

numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]

**Key differences:**

- map() applies a function to each item of an iterable and returns a list of results.
- reduce() applies a function to the items of an iterable, reducing it to a single output.
- filter() constructs an iterator from elements of an iterable for which a function returns true.
